```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

df = pd.DataFrame() #blank dataframe
df['rand'] = np.random.rand(100)  #dataframe which is 100 random numbers between 0 and 1
df['rand_sqrt'] = df['rand']**0.5

# line graphs
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.plot(df['rand'], color='blue')

plt.subplot(1,2,2)
#plt.plot(df['rand_sqrt'], color='green')
plt.scatter(x=df['rand'], y=df['rand_sqrt'], color='green')

plt.show()

# bar graphs
objects = ('Wellesley', 'Smith', 'Mt. Holyoke', 'Vassar', 'Barnard', 'Bryn Mawr', 'Radcliffe')
y_pos = np.arange(len(objects))
num = [10,8,6,4,2,9,7]
 
plt.bar(y_pos, num, align='center', alpha=0.9)
plt.xticks(y_pos, objects)
plt.ylabel('Attendance')
plt.title('Seven Sisters Attendance')
 


plt.show()
```
