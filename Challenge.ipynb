```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('/Users/rachellam/Downloads/UN_Adult_lit_rate.csv')

#bar charts
plt.subplot(1, 2, 1)
df1 = df.iloc[6:8]
#df1.plot(x='Year', y='Value', kind='bar', legend=False, color='blue') ####
#plt.plot(x=df1['Year'], y=df1['Value'], kind='bar', legend=False, color='blue')
plt.title('Adult Literacy Rate in Algeria, Male 15+')

plt.subplot(1, 2, 2)
df2 = df.iloc[4:6]
#df2.plot(x='Year', y='Value', kind='bar', legend=False, color='pink')
plt.plot((x=df2['Year'], y=df2['Value'], kind='bar', legend=False, color='pink')
plt.title('Adult Literacy Rate in Algeria, Female 15+')


## This shows a fairly dramatic increase in adult literacy rates among both sexes in Algeria between 1987 and 2002.
## The increase could be attributed to improved educational levels, a more stable government, or other factors.

df3 = df.iloc[12:14]
df3.plot(x='Year', y='Value', kind='bar', legend=False, color='blue')
plt.title('Adult Literacy Rate in Argentina, Male 15+')


df4 = df.iloc[10:12]
df4.plot(x='Year', y='Value', kind='bar', legend=False, color='pink')
plt.title('Adult Literacy Rate in Argentina, Female 15+')


#scatter plots

df1.plot.scatter(x='Year', y='Value', color='blue')
plt.title('Adult Literacy Rate in Algeria, Male 15+')


df2.plot.scatter(x='Year', y='Value', color='pink')
plt.title('Adult Literacy Rate in Algeria, Female 15+')


df3.plot.scatter(x='Year', y='Value', color='blue')
plt.title('Adult Literacy Rate in Argentina, Male 15+')


df4.plot.scatter(x='Year', y='Value', color='pink')
plt.title('Adult Literacy Rate in Argentina, Female 15+')


#line graphs

df1.plot(x='Year', y='Value', legend=False, color='blue')
plt.title('Adult Literacy Rate in Algeria, Male 15+')

df2.plot(x='Year', y='Value', legend=False, color='pink')
plt.title('Adult Literacy Rate in Algeria, Female 15+')

df3.plot(x='Year', y='Value', legend=False, color='blue')
plt.title('Adult Literacy Rate in Argentina, Male 15+')

df4.plot(x='Year', y='Value', legend=False, color='pink')
plt.title('Adult Literacy Rate in Argentina, Female 15+')


plt.show()

## Between 1991 and 2001, there was a much smaller jump in literacy rates in Argentina, since its people were already 
## fairly educated. It seems that educational levels are slow to decrease despite major economic recessions.
```
